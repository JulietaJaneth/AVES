Ahora extraigamos la información de los xml:

````python
ID=[]
Quality=[]
FileName=[]
ClassID=[]
Date=[]
Time=[]
Country=[]
Locality=[]
Latitude=[]
Longitude=[]
Elevation=[]
Author=[]
AuthorID=[]
Content=[]
Year=[]
Family=[]
Genus=[]
Species=[]
VernacularNames=[]
BackgroundSpecies=[]

ruta2= '/media/birds/data/LIFECLEF2017_BIRD_XC_WAV_RN*.xml'

for name in sorted(glob.glob(ruta2)):
    archivo_xml=ET.parse(name)
    Quality.append(archivo_xml.findtext('Quality'))
    ID.append(archivo_xml.findtext('MediaId'))
    FileName.append(archivo_xml.findtext('FileName'))
    ClassID.append(archivo_xml.findtext('ClassId'))
    Date.append(archivo_xml.findtext('Date'))
    Time.append(archivo_xml.findtext('Time'))
    Country.append(archivo_xml.findtext('Country'))
    Locality.append(archivo_xml.findtext('Locality'))
    Latitude.append(archivo_xml.findtext('Latitude'))
    Longitude.append(archivo_xml.findtext('Longitude'))
    Elevation.append(archivo_xml.findtext('Elevation'))
    Author.append(archivo_xml.findtext('Author'))
    AuthorID.append(archivo_xml.findtext('AuthorID'))
    Content.append(archivo_xml.findtext('Content'))
    Year.append(archivo_xml.findtext('Year'))
    Family.append(archivo_xml.findtext('Family'))
    Genus.append(archivo_xml.findtext('Genus'))
    Species.append(archivo_xml.findtext('Species'))
    VernacularNames.append(archivo_xml.findtext('VernacularNames'))
    BackgroundSpecies.append(archivo_xml.findtext('BackgroundSpecies'))
````

Ahora lo que sigue es procesar lo datos donde calcularemos la frecuencia de muestreo, número de muestras y la dulación de cada archivo de audio. Además de que le sacaremos la transformada de fourier para poder obtener el esprecrograma así como también pondremos su amplitud en decibeles, guardarémos tanto el espectrograma como el espectrograma en decibeles en el disco duro con ayuda de la librería pickle.

Espectrograma estará en '/media/birds/Espectrogramas'.

Y el espectrograma en decibeles en '/media/birds/Espectrogramas_dB'

## Consejo

Si tienes dudas de estos conceptos te recomiendo el siguiente blog.
https://l52mas.gitlab.io/posts/librosa-report/

````python
sampling_rates=[]
Muestras=[]
Duracion=[]
for i in df_divididor['FileName']:
    
    audio ='/media/birds/data/'+i
    
    y, sampling_rate = librosa.load(audio,sr=None)
    sampling_rates.append(sampling_rate* 1e-3)
    
    T = y.size / sampling_rate
    T2="{0:.3f}".format(T)
    dt = 1 / sampling_rate
    t = np.r_[0:T:dt]
    Muestras.append(y.size)
    Duracion.append(T2)
    X = librosa.stft(y)
    X_dB = librosa.amplitude_to_db(np.abs(X))
    f=open(('/media/birds/Espectrogramas/'+i).replace(".wav", ".pkl"), 'wb')
    pickle.dump(X, f)
    f.close()
    f1=open(('/media/birds/Espectrogramas_dB/'+'dB_'+i).replace(".wav", ".pkl"), 'wb')
    pickle.dump(X_dB, f1)
    f1.close()
````

Agregemos lo calculado al dataframe.

````python
df_tiempos= pd.DataFrame({'sampling rate (kHz)':sampling_rates,'No.Muestras':Muestras,'Duración (s)':Duracion})
df = pd.concat([df_divididor, df_tiempos], axis=1)
df
````

Ahora que el dataframe está completo, almacenemoslo en el disco duro con pickle.

````python
f2=open('/media/birds/BirdsDataFrame.pkl', 'wb')
pickle.dump(df, f2)
f2.close()
````
